What was the average (mean) TotalPay of all employees per year?

How many unique job titles are there?

```python
salaries['JobTitle'].nunique()
```

What are the top 5 most common jobs?

```python
salaries['JobTitle'].value_counts()[:5]
```

How many people have the word **Chief** in their job title?



```python
mask = salaries['JobTitle'].apply(lambda txt:  True if 'chief'in txt.lower() else False)
salaries[mask]['JobTitle'].count()
```
